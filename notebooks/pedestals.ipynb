```python
# histogram of pedestal levels
h_pedestal = ROOT.TH1F("h_pedestal", "Pedestal Levels;Pedestal Level;Counts", 2000, -2000, 0)
h_pedestalStdev = ROOT.TH1F("h_pedestalStdev", "Pedestal Stdev;Pedestal Stdev;Counts", 100, 0, 100)
h_nPedestalSamples = ROOT.TH1F("h_nPedestalSamples", "Number of Pedestal Samples;Number of Samples;Counts", 100, 0, 100)
h_pedestalStartSample = ROOT.TH1F("h_pedestalStartSample", "Pedestal Start Sample;Start Sample;Counts", 1000, 0, 1000)
```

```python
#Reset the histograms
h_pedestal.Reset()
h_pedestalStdev.Reset()
h_nPedestalSamples.Reset()
h_pedestalStartSample.Reset()

# Loop over the tree
for entry in tree:
    waveforms = entry.pedestal_waveforms #grouped_waveformsRF
    for waveform in waveforms:
        h_pedestal.Fill(waveform.pedestalLevel)
        h_pedestalStdev.Fill(waveform.pedestalStdev)
        h_nPedestalSamples.Fill(len(waveform.pedestalSamples))
        h_pedestalStartSample.Fill(waveform.pedestalStartSample)
```

```python
c = ROOT.TCanvas("c", "Pedestal Levels", 1200, 800)
c.Divide(2,2)

c.cd(1)
h_pedestal.Draw()

c.cd(2)
h_pedestalStdev.Draw()
ROOT.gPad.SetLogy()

c.cd(3)
h_nPedestalSamples.Draw()
ROOT.gPad.SetLogy()

c.cd(4)
h_pedestalStartSample.Draw()
ROOT.gPad.SetLogx()
ROOT.gPad.SetLogy()

c.Draw()
```
